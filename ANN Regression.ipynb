```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
tf.__version__
```

```python
dataset = pd.read_excel('/content/Folds5x2_pp.xlsx')
```

```python
x = dataset.iloc[: , :-1].values
y = dataset.iloc[: , -1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state = 0)
```

```python
ann = tf.keras.models.Sequential() # ANN structure and input layer
```

```python
ann.add(tf.keras.layers.Dense(units = 6, activation = 'relu' )) # First hidden layer
ann.add(tf.keras.layers.Dense(units = 6, activation = 'relu' )) # Second hidden layer
ann.add(tf.keras.layers.Dense(units = 1)) # OUTPUT layer
```

```python
ann.compile(optimizer ='adam' , loss = 'mean_squared_error') # COMPILATION
```

```python
ann.fit(x_train,y_train,batch_size = 32 , epochs = 100) # batch_size = 32 because its a default/classic value
```

```python
y_pred = ann.predict(x_test)
y_pred = (y_pred > 0.5) # take probabilities in the form of either 0 or 1.
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
