```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
tf.__version__
```

```python
dataset = pd.read_csv('/content/Churn_Modelling.csv')
```

```python
x = dataset.iloc[: , 3:-1].values
y = dataset.iloc[: , -1].values
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x[: , 2] = le.fit_transform(x[: , 2])
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers = [('encoder' , OneHotEncoder() , [1])], remainder = 'passthrough')
x = np.array(ct.fit_transform(x))
```

```python
print(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x_train = ss.fit_transform(x_train)
x_test = ss.transform(x_test)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
ann = tf.keras.models.Sequential() # ANN structure and input layer
```

```python
ann.add(tf.keras.layers.Dense(units = 6, activation = 'relu' )) # First hidden layer
ann.add(tf.keras.layers.Dense(units = 6, activation = 'relu' )) # Second hidden layer
ann.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid' )) # OUTPUT layer
```

```python
ann.compile(optimizer ='adam' , loss = 'binary_crossentropy', metrics = ['accuracy']) # COMPILATION
```

```python
ann.fit(x_train,y_train,batch_size = 32 , epochs = 100) # batch_size = 32 because its a default/classic value
```

```python
print(ann.predict(ss.transform([[1,0,0,600,1,40,3,60000,2,1,1,50000]])))
```

```python
y_pred = ann.predict(x_test)
y_pred = (y_pred > 0.5) # take probabilities in the form of either 0 or 1.
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python
accuracy_score(y_test,y_pred)
```
